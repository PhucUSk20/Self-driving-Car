```python
import serial
import time

# Thiết lập cổng Serial và tốc độ baud
arduino_port = '/dev/ttyACM1'  # Thay thế bằng cổng thực tế của bạn
baud_rate = 9600  # Tốc độ baud phù hợp với thiết lập trên Arduino

# Tạo đối tượng Serial
ser = serial.Serial(arduino_port, baud_rate, timeout=1)

# Đợi một chút để đảm bảo kết nối ổn định
time.sleep(2)

# Đọc dữ liệu từ Arduino
try:
    while True:
        if ser.in_waiting > 0:
            line = ser.readline().decode('utf-8').rstrip()
            print(line)
            # Bạn có thể xử lý dữ liệu tại đây hoặc lưu trữ nó vào danh sách, tệp, v.v.
except KeyboardInterrupt:
    print("Kết thúc nhận dữ liệu")

# Đóng cổng Serial khi hoàn thành
ser.close()
```
