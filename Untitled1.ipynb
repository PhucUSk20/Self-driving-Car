```python
import serial
import time

# Thit lp cng Serial v tc  baud
arduino_port = '/dev/ttyACM1'  # Thay th bng cng thc t ca bn
baud_rate = 9600  # Tc  baud ph hp vi thit lp trn Arduino

# To i tng Serial
ser = serial.Serial(arduino_port, baud_rate, timeout=1)

# i mt cht  m bo kt ni n nh
time.sleep(2)

# c d liu t Arduino
try:
    while True:
        if ser.in_waiting > 0:
            line = ser.readline().decode('utf-8').rstrip()
            print(line)
            # Bn c th x l d liu ti y hoc lu tr n vo danh sch, tp, v.v.
except KeyboardInterrupt:
    print("Kt thc nhn d liu")

# ng cng Serial khi hon thnh
ser.close()
```
